```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
df = sns.load_dataset('titanic')
```

```python
df.head()
```

```python
sns.countplot(x='sex', data=df)
plt.title('Gender Distribution')
plt.show()
```

```python
sns.countplot(x='class', data=df)
plt.title('Class Distribution')
plt.show()
```

```python
sns.histplot(df['age'].dropna(), bins=30, kde=True)
plt.title('Age Distribution')
plt.show()
```

```python
g = sns.FacetGrid(df, col='class', row='sex', margin_titles=True)
g.map(plt.hist, 'age', bins=30, color='steelblue')
plt.show()
```

```python
sns.heatmap(df.isnull(), cmap='viridis', cbar=False)
plt.title('Missing data map')
plt.show()
```

```python
sns.scatterplot(x='age', y='fare', data=df)
plt.title('The connection between age and fare')
plt.show()
```

```python
sns.scatterplot(x='class', y='age', data=df, hue='survived', palette='Set2')
plt.title('The connection between age and class')
plt.show()
```
