```python
# map() metodu bir series ve ya sutundaki her elemente muyyen bir emeliyyati tetbiq etmek ucun istifade olunur.
# Onun temel istifade qaydasi bu sekildedir: new_elements = old_elements.map(function)
# Burada 'old_elements' deyisdirilecek olan orjinal elementleri temsil edir.
# 'function' ise her bir elemente tetbiq olunucaq deyisdirme emeliyyatini temsil edir.
# Tutaq ki, biz her bir serabin aldigi qiymeti 0-a endirmek isteyirik bunu asagidaki sekilde ede bilerik.
# map() icerisinde istifade etdiyimiz funksiya bir argument almalidir ve hemin argumentin deyisdirilmis versiyasini qaytarmalidir
# map() butun elementlerin deyisdirildiyi yeni bir seriya qaytarir.

points_mean = df.points.mean()
df.points.map(lambda x: x - points_mean)
```

```python
# Eger biz butun DataFrame i deysidirmek istiyirikse bunun ucun apply() metodundan istifade etmek daha meqsede uygun olar.
# yuxarida etdiyimiz emeliyyati butun DataFrame apply() metodu ile asagidaki kimi tetbiq ede bilerik

def remean_points(row):
    row.points = row.points - points_mean
    return row

df.apply(remean_points, axis='columns')

# Eger burada axis='index' yazsaydiq bu zaman funksiyaya setir gondermek evezine sutun gondermeli idik.
```
