```python
# 23. Yeni bir sütun yaradın əgər ölkənin adı “Canada”-dırsa sətirdə 3, 
# əgər points 95-dən böyük bərabərdirsə sətridə 3, əgər 85-dən böyük bərabər və 95-dən kiçikdirsə sətirdə 2, digər bütün hallarda 1 yazılsın.

def star_ratings(row):
    if row.country == 'Canada':
        return 3
    elif row.points >= 95:
        return 3
    elif row.points >= 85:
        return 2
    else:
        return 1

star = pd.Series(df.apply(star_ratings, axis='columns'))

df['star ratings'] = star

df.sample(10)
```
