```python
# drop() metodu ile dataset icerisinden bize lazim olmayacaq sutunlari ve ya setireri sile bilerik.
# Sutunlari silmek ucun columns=[] adinda bir list yaziriq ve bu listin icerisine silmek istediyimiz sutunlarin adlarini qeyd edirik
# setirleri silmek ucun ise sadece olaraq silmek istediyimiz setrin index nomresini qeyd edirik
# Asagidaki numunede 'region_2' ve 'taster_twitter_handle' sutunlarini sileceyik
# Lakin bir seyi yadda saxlamaq lazimdirki drop() metodu orjinal dataframe uzerinde deyisiklik etmir. Uzerinde deyisiklik edilmis yeni bir dataframe qaytarir.
# Eger biz orjinal dataframe uzerinde deyisiklik etmek isteyirikse bu zaman drop() metodu icerisinde 'inplace=True' ifadesini isletmeliyik. Mes: drop(columns='region_2', inplace=True)
 
df.drop(columns=['region_2', 'taster_twitter_handle'])
```

```python
# query() metdou ile dataframe uzerinde filerleme emliyyatlari ve ya istediyiniz sertlere uygun sutunlari ala bilersiniz. 
# Asagidaki numunede 'price > points' serti bize pricein deyerinin pointsin deyerinden boyuk oldugu setirleri qaytarir
# Daha qarmasiq sertler yazamaq ucunde and ve ya or kimi metniq ifadelerinde istifade ed bilersiniz. Mes: df.query("price > points and country = 'US'")

df.query('price > points')
```
