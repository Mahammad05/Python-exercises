```python
df1=pd.DataFrame({'nomre':[12, 10, 24, 4, 27],
    'ad':['Ulviyye', 'Gunay', 'Seymur', 'Aysel', 'Emin']})

df2=pd.DataFrame({"nomre":[12, 12, 10, 4, 4, 45],
                 'ders':['Python', 'SQL', 'SQL', 'Dil', 'Cebr', 'SQL'],
                 'qiymet': [47, 50, 57, 78, 75, 10]})
```

```python
df1
```

```python
df2
```

```python
df1.merge(df2, how='inner', on='nomre')
```

```python
df1.merge(df2, how='left', on='nomre')
```

```python
df1.merge(df2, how='right', on='nomre')
```

```python
df1.merge(df2, left_on='nomre', right_on='qiymet')
```

```python
df1.merge(df2, left_on='nomre', right_on='ders')
```
